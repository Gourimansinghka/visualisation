```python
plt.style.use('classic')
plt.figure(figsize=(15,5))
x=np.array([1,2,3,4,5,6])
y=np.power(x,3)
plt.plot(x,y)
plt.title('my first graph')
plt.xlabel('it is X-axis')
plt.ylabel('it is Y-axis')
plt.show()
```

```python
x=np.linspace(0,10,1000)
y1=np.sin(x)
y2=np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

# seaborn

 # 1.Categorical plot ------- At least one categorical col
     i)count plot
     ii)box plot
     iii)bar plot
     iv)violin plot
     v)swarm plot
 # 2.Distribution plot ------ At least one continous numercical
     i)dist plot
     ii)histogram
     iii)kbe plot
     iv)line plot
     v)reg plot
     vi)join plot
# 3.Matrix --------- cont. numerical or categorical(numeric)
     i)pair plot
     ii)heatmap    
# 4.Regression plot -------- two continous  column
     i)scatter plot
     ii)regplot
